```python
# Davies Bouldin Method Araştırılıp Uygulanacak.
```

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
import matplotlib.pyplot as plt
```

```python
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
```

```python
kmeans = KMeans(n_clusters=4, random_state=0)
labels = kmeans.fit_predict(X)
```

```python
db_index = davies_bouldin_score(X, labels)
print("Davies-Bouldin Index:", db_index)
```

```python
# Scaling (Özellik Ölçekleme) nedir? Nasıl yapılır?
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
iris = load_iris()
X = iris.data
feature_names = iris.feature_names
```

```python
minmax_scaler = MinMaxScaler()
X_minmax = minmax_scaler.fit_transform(X)
```

```python
plt.figure(figsize=(8, 4))
plt.title("Min-Max ile Ölçeklenmiş Veriler")
for i in range(X_minmax.shape[1]):
    plt.plot(X_minmax[:, i], label=feature_names[i])
plt.legend()
plt.xlabel("Örnek Index")
plt.ylabel("Ölçeklenmiş Değer")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
std_scaler = StandardScaler()
X_std = std_scaler.fit_transform(X)
```

```python
plt.figure(figsize=(8, 4))
plt.title("Z-Score ile Ölçeklenmiş Veriler")
for i in range(X_std.shape[1]):
    plt.plot(X_std[:, i], label=feature_names[i])
plt.legend()
plt.xlabel("Örnek Index")
plt.ylabel("Ölçeklenmiş Değer")
plt.grid(True)
plt.tight_layout()
plt.show()
```
